# guessing game

```python
import random
```

```python

jackpot = random.randint(1,100)
guess = int(input("guess kro"))
Counter =1
while guess!=jackpot:
  if guess < jackpot:
    print("guess higher")
  else:
    print("lower guess")
  guess = int(input("guess kro"))
  Counter+=1
else:
  print("correct guess")
  print("attempt",Counter)

```

```python
jack = random.randint(1,100)
guess = int(input("guess kro"))
count = 1
while guess!=jack:
  if guess<jack:
    print("guess higher")
  else:
    print("guess lower")
  guess = int(input("guess kro"))
  count+=1
else:
  print("correct guess")
  print("attempt",count)
```
